```python
A = (5, 3)
B = (2, 4)
```

```python
from scipy.spatial import distance
```

```python
# Euclidean Distance
euclidean_dist = distance.euclidean(A, B)
euclidean_dist
```

```python
# Manhattan Distance
manhattan_dist = distance.cityblock(A, B)
manhattan_dist
```

```python
# Cosine Distance
cosine_dist = distance.cosine(A, B)
cosine_dist
```
