```python
import pandas as pd
data = pd.DataFrame(data = {
    'X': [0, 1.1, 1, 2, 2, 4, 5, 5],
    'Y': [0, 1.5, 4, 2, 3, 1, 0, 4]
})

data
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['X'], data['Y'])

numbers = [0, 1, 2, 3, 4, 5, 6, 7]
for index, value in enumerate(numbers):
    plt.annotate(value, (data['X'][index], data['Y'][index]), size = 12)
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
plt.title('Dendrogram')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=2, metric = 'euclidean',  linkage='ward')
pred = model.fit_predict(data)
```

```python
pred
```

```python
data['Cluster'] = pred
data
```

```python
cluster1 = data[data['Cluster'] == 0]
cluster2 = data[data['Cluster'] == 1]

plt.scatter(cluster1['X'], cluster1['Y'], color = 'red')
plt.scatter(cluster2['X'], cluster2['Y'], color = 'blue')
```
