Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 123th index, we can make a new column based on the Region.

1 : "Bejaia Region Dataset"
2 : "Sidi-Bel Abbes Region Dataset"

Add new column with eegion

```python
df = dataset

df.loc[:123, "Region"] = 0
df.loc[123:, "Region"] = 1

df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Change the required columns as integer type

```python
df[['day','month', 'year', 'Temperature', 'RH','Ws']] = df[['day','month', 'year', 'Temperature', 'RH','Ws']].astype(int)
```

```python
df.info()
```

### Changing other columns to float type

```python
objects = df.select_dtypes(include='object').columns.tolist()
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```
