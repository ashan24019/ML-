```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
dataset
```

```python
dataset.info()
```

Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 123th index, we can make a new column based on the Region.

1 : "Bejaia Region Dataset"
2 : "Sidi-Bel Abbes Region Dataset"

Add new column with eegion

```python
df = dataset

df.loc[:123, "Region"] = 0
df.loc[123:, "Region"] = 1

df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Change the required columns as integer type

```python
df[['day','month', 'year', 'Temperature', 'RH','Ws']] = df[['day','month', 'year', 'Temperature', 'RH','Ws']].astype(int)
```

```python
df.info()
```

### Changing other columns to float type

```python
objects = df.select_dtypes(include='object').columns.tolist()
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
## Save the cleaned dataset
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv', index=False)
```

## Exploratory Data Analysis

```python
df_copy = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes']= df_copy['Classes'].str.strip().str.lower()
df_copy['Classes'].value_counts()
```

```python
# Encoding Categories in Classes
df_copy['Classes'] = df_copy['Classes'].map({'not fire':0, 'fire':1})
df_copy['Classes'].value_counts()
```

```python
df_copy
```

```python
## Plot density plot for all features

plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
## percentage of Pie Chart

percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
#plot the pie chart
classLabels = ['Fire', 'Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classLabels, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'salmon'])
plt.title('Percentage of Classes')
plt.show()
```

### Monthly Fire Analysis

```python
df['Classes']= df['Classes'].str.strip().str.lower()
df['Classes'] = df['Classes'].map({'not fire':0, 'fire':1})
```

```python
dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=dftemp)
plt.title('Monthly Fire Analysis in Sidi-Bel Abbes Region', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Fires', fontsize=14)
```

```python
dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=dftemp)
plt.title('Monthly Fire Analysis in Bejaia Region', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Fires', fontsize=14)
```
